# Pokemon Picture Extraction

```python
url = 'https://pokemondb.net/pokedex/national'
data = requests.get(url)
soup = BeautifulSoup (data.text, 'html.parser')

storage_img = []
for img in soup.find_all('span',{'class':'img-fixed img-sprite'}):
    storage_img.append(img['data-src'])
```

# Mapping 

Mapping dititikberatkan pada area San Fransisco Bay

```python
m = folium.Map(
location = [37.801481, -122.460002],
tiles='OpenStreetMap',
zoom_start = 13
)

for i in range(0, 1501):
    num = random.randint(0,len(df))
    folium.Marker(
    [df.iloc[num]['lat'], df.iloc[num]['lng']],
    popup = f'{df.iloc[num]["name"]}',
    tooltip =  f'<img width="100px" src="https://img.pokemondb.net/sprites/sword-shield/icon/{df.iloc[num]["name"].lower()}.png"/><center>{df.iloc[num]["name"]}</center>',
    icon = folium.features.CustomIcon(f'https://img.pokemondb.net/sprites/sword-shield/icon/{df.iloc[num]["name"].lower()}.png', icon_size=(50,50))).add_to(m)

m
```
